## Business Understading Phase

The adult dataset is census data extracted from census bureau database website (http://www.census.gov/ftp/pub/DES/www/welcome.html). The data was extracted and donated by Barry Becker and Ronny Kohavi, 1996. There are 15 attributes, 6 continuous and 9 categorical. One of the categorical attributes, "Income" is a binary class which tells whether the person earns an income of more than 50K or not. By choosing this as the dependent variables, makes this dataset a good candidate for a binary classification problem: Classifiy entities into earning more than 50K (">50K") or not ("<=50K").

Since the release, it has been used as a benchmarking dataset (especially in the 2000s). Recently a Kaggle post (https://www.kaggle.com/mattmet/income-prediction-xgboost-accuracy-86-02/comments) gave a test accuracy of 86% using XGBoost, which is an improvement over the originally reported scores. Just to see the performance on the unprocessed data, a random forest classifier with 1000 predictors and max depth of 2 was fed with the training data (adult.data). The model achieved a score of 77.6% on test data which is not too bad given that this was unprocessed data with missing values that added up as noise. For this assignment, we're only using the training data file ('adult.data') which contains 32561 entities.

The dependent attribute, "Income" has an unsymmetric distribution. The class '<=50K' has 24720 entities whereas '>50K' has only 7841.

## Data Understanding Phase

### a) In this section we go through a high-level overview of all the attributes:

* Age: is a continuous attribute that. as the name says, is age information of each entity.
* Workclass: is a categorical attribute that tells about the job class of each entity, eg. private, government, with-pay etc.
* Final Weight: is a continuous attribute, which characterizes state based attributes of entities.
* Education: is a categorical attribute that contains educational levels like Bachelor, Ph.D. etc.
* Education Num: is same as Education, but the integer values represent the order of the classes. For example, Bachelor has a value of 13 whereas Master is 14.
* Marital Status: tells about the marital status of entities, eg. Divorced, Never married etc.
* Occupation: is a categorical attribute. Each class is the occupation of an entity.
* Relationship: gives the relationship which that person belongs to in his/her family.
* Race: tells whether the person is White etc.
* Sex: gives the gender of the person.
* Capital Gain: is a continuous attribute that tells about the capital gains of entities.
* Capital Loss: tells about the capital losses.
* Hours Per Week: Number of hours the person works for per week.
* Native Country: is the native country of the entity.

The function \_3plot defined below takes 3 arguments (attribute names) and returns for scatter plots. The first 3 plots are 2D scatter plots for all the 3 attribute combinations. The last plot is a 3D scatter plot with the three attributes on the three axes.
```python
from mpl_toolkits.mplot3d import axes3d, Axes3D
def _3plot(a1,a2,a3):

 a = [] ; b = [] ; c = []
 A = [] ; B = [] ;  C = []

 for i in range(len(train)):
  if train[14].iloc[i]== ' >50K':
   a.append(train[a1].iloc[i])
   b.append(train[a2].iloc[i])
   c.append(train[a3].iloc[i])
  else:
   A.append(train[a1].iloc[i])
   B.append(train[a2].iloc[i])
   C.append(train[a3].iloc[i])

 plt.scatter(a,b,c = 'red',label ='>50K')
 plt.scatter(A,B,c = 'blue',label ='<=50K')
 plt.xlabel(data_keys[a1])
 plt.ylabel(data_keys[a2])
 plt.legend()
 plt.show()

 plt.scatter(b,c,c = 'red',label ='>50K')
 plt.scatter(B,C,c = 'blue',label ='<=50K')
 plt.xlabel(data_keys[a2])
 plt.ylabel(data_keys[a3])
 plt.legend()
 plt.show()

 plt.scatter(a,c,c = 'red',label ='>50K')
 plt.scatter(A,C,c = 'blue',label ='<=50K')
 plt.xlabel(data_keys[a1])
 plt.ylabel(data_keys[a3])
 plt.legend()
 plt.show()

 points = len(train)
 fig = plt.figure()
 ax = fig.add_subplot(111, projection='3d')
 ax.scatter(a[:points],b[:points],c[:points],c = 'red',label ='>50K')
 ax.scatter(A[:points],B[:points],C[:points],c = 'blue',label ='<=50K')
 plt.xlabel(data_keys[a1])
 plt.ylabel(data_keys[a2])
 plt.legend()

 plt.show()
```

## Contingency Tables (Categorical Data)

The categorical data has no order to it so the only way to analyze is by making 2 (or more) attribute contingency tables. The function cont below computes the contingency table for two variables and also runs a chi-square test on the computed table to find dependence/independence of attribute pairs.

```python
from scipy.stats import chi2_contingency
from scipy.stats import chi2
def cont(table):
    #adapted from https://machinelearningmastery.com/chi-squared-test-for-machine-learning/
    table = np.array(table)
    stat, p, dof, expected = chi2_contingency(table)
    print('dof=%d' % dof)
    
    prob = 0.95
    critical = chi2.ppf(prob, dof)
    print('probability=%.3f, critical=%.3f, stat=%.3f' % (prob, critical, stat))
    if abs(stat) >= critical:
        print('Dependent (reject H0)')
    else:
        print('Independent (fail to reject H0)')
    # interpret p-value
    alpha = 1.0 - prob
    print('significance=%.3f, p=%.3f' % (alpha, p))
    if p <= alpha:
        print('Dependent (reject H0)')
    else:
        print('Independent (fail to reject H0)')
```

## Data Processing Phase

### a)

* We saw above that Education Num is same as Education with added order to the classes. This makes Education a redundant attribute so we can remove it.
* The correlation table shows that fnlwgt has no correlation with Income or any other continuous attributes so we can remove it. 
* Capital Gain and Capital Loss can be combined to form a single attribute.

So we delete **Education Num**, **Fnlwgt**, **Capital Gain** and **Capital Loss**. 

### b)

* We replace the categorical attributes by continuous ones. In order to do this, we replace each class with integers from the set [0, N] where N is the number of classes in an attribute.
* We add the mode for each attributes wherever the values are missing (?)

### c)

As stated in part a), we add a new attribute **Capital Gain 2**. This attribute has values from **Capital Gain** and negative of the values from **Capital Loss**.

### d)

In the code snippet below we perform the changes (except changing categorical to continous, as it does not affect performance) that we discussed above. The randomforest classifier recieved a training accuracy of 78.5-79.5% which is an improvement over the unprocessed data (77.6%).

The second snippet shows \_3plot result for attributes **Age**, **Capital Gain 2** and **hours per week**.

The code generates a new file called 'adult.data.new'. This is the final dataframe produced at the end of the last snippet.

```python
new_keys = list(data_keys.keys())
new_train = train
new_keys.remove(2)
new_keys.remove(4)
new_train[10] = train[10]-train[11] #new attribute Capital Gain 2
new_keys.remove(11)
new_train = new_train[new_keys]
data_keys={0: 'age',1: 'workclass',2: 'fnlwgt',3: 'education',4: 'education-num',5: 'marital-status',6: 'occupation' ,7: 'relationship' ,8: 'race' ,9: 'sex' ,10: 'capital-gain-2' ,11: 'capital-loss',12: 'hours-per-week',13: 'native-country',14: 'income'}

#Replacing ? with mode values
for i in [1,6,13]:
    mode = new_train[i].describe()['top']
    new_train[i] = new_train[i].replace(' ?',mode)

new_train.to_csv('adult.data.new',header = ['age','workclass', 'education-num','marital-status', 'occupation' , 'relationship' ,'race' , 'sex' ,'capital-gain-2' , 'hours-per-week', 'native-country','income'])


new_train= np.array(new_train)

for i in range(len(new_train)):
    new_train[i][11] = new_train[i][11].replace('.','')
    
tot=new_train

for i in range(tot.shape[1]):
    temp= list(set(tot[:,i]))
    dic = dict()
    for j in range(len(temp)):
        dic.update({temp[j]:j})
    for j in range(len(new_train)):
        new_train[j][i] = dic[new_train[j][i]]
        
#Random Forest Classifier        
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(n_estimators=100, max_depth=2, random_state=0)

X = new_train[:,:-1]
Y = new_train[:,-1]
Y = Y.astype('int')

clf.fit(X,Y)

from sklearn.metrics import accuracy_score

print('Train Accuracy ' + str(accuracy_score(clf.predict(X),Y)))
```
